```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris.xlsx")
```

```python
data.head()
```

```python
x = data.iloc[:,:-1]
```

```python
y = data.iloc[:,-1]
```

```python
y
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
model.predict([[5, 2, 3.5, 1]])
```

```python
data2 = pd.read_excel("iris_test.xlsx")
```
